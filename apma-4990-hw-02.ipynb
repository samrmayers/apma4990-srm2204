and compute $Ax$.

```python
#CODE HERE
A = np.array([[1,2,3],[4,5,6]])
x = np.array([[1],[-1],[2]])
A@x
```

Construct a function that multipies a matrix times a vector

```python
import numpy as np
def matrix_vector_mult(A,x):
    """
    This function returns the product of the matrix A and vector x.
    The product is computed by looping over columns only.
    There is no check that the the number of columns of A is equal to the number of elements of x
    """
    nrows, ncolumns = A.shape
    #CODE HERE
    b = [0]*A.shape[0]
    if ncolumns == len(x):
        for ii in range(0,nrows):
            b[ii] = int(np.dot(A[ii],x))
        return np.array(b)
    else:
        return NaN
matrix_vector_mult(A,x)
```

### 3.2 Compute the climatology and anomaly

Compute the climatology of Nino 3.4 in 3 different ways and get 3 different answers. 

1. Which one is correct? 
    Not Method 1, Not Method 2, so Method 3 is correct
2. Explain the errors 
    
    Method 1 - 
        The problem here is that the line "months = range(imonth, 12, nmonths)" is 
        calculating the range from imonth to 12, with a step size of nmonths=492. This is 
        backwards, since the step size should be 12 and the upper bound 492.
        
    Method 2 - 
        Method 2's mistake is that the values occur in chronological order, so the way the 
        data is being reshaped is incorrect since the columns being averaged do not contain 
        data from the same month. Each row size should be 12 (aka 12 columns), but this 
        code is producing 12 rows, not columns. If each row was size 12, then element 1 in 
        the row would be January for all rows, element 2 February, etc. In order for each
        row to be size 12, there must be 41 rows for this data set.
3. Fix the errors in the other two.

```python
# 41 years
nmonths = n34.size
nmonths/12
```

Method 1 (loops)

```python
climatology = np.full(12,np.nan)
anomaly = np.zeros_like(n34)
```

```python
for imonth in range(1,12):
    months = range(imonth, nmonths, 12)
    climatology[imonth] = n34[months].mean()
    anomaly[months] = n34[months] - n34[months].mean() #CODE
print(climatology)
```

Method 2 (reshape)

```python
# reshape the data so that each of the 41 columns contains a year and then average over columns (years)
climatology2 = n34.values.reshape(41,-1).mean(axis=0)
anomaly2 = n34.values.reshape(41,-1) - climatology
anomaly2.resize(nmonths)
print(climatology2)
```

Method 3 (xarray groupby)  
http://xarray.pydata.org/en/stable/groupby.html

```python
climatology3 = n34.groupby("T.month").mean()
anomaly3 = n34.groupby("T.month") - climatology3
print(climatology3)
```

```python
# plot them
plt.plot(climatology,label = 'loops')
plt.plot(climatology2,label = 'reshape')
plt.plot(climatology3, label = 'groupby')
plt.legend()
plt.xticks(np.arange(0,12), m_str)
plt.title('Nino 3.4 climatology');
```
